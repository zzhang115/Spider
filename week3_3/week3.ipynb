```python
import pymongo
```

```python
client = pymongo.MongoClient('localhost', 27017)
crime = client['crime']
crime_info = crime['crime_info']
```

```python
for i in crime_info.find({'state' : {'$in' : ['new_york', 'oregon']}},{'state' : 1, 'city' : 1, 'population' : 1, '_id' : 0}):
    print(i)
```

```python
for i in crime_info.find({'state' : {'$in' : ['california']}},{'state' : 1, 'murder' : 1, 'city' : 1, 'population' : 1, '_id' : 0}):
    print(i)
```
